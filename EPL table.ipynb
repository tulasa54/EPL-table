```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
EPL_url = 'https://www.espn.com/soccer/standings/_/league/eng.1'
EPL_response = requests.get(EPL_url)
EPL_soup = BeautifulSoup(EPL_response.content, 'html.parser')
```

```python
clubs_list = []
games_data_list = []

club_tags = EPL_soup.find_all('span', class_ = "hide-mobile")

for club_names in club_tags:
    clubs_list.append(club_names.get_text())
    
games_data_tags = EPL_soup.find_all('tbody', class_ = "Table__TBODY")
for games_data in games_data_tags:
    games_data = games_data.find_all('tr', class_ = 'Table__TR')
    for games_info in games_data: 
        games_list = []
        games_info = games_info.find_all('span', class_ = 'stat-cell')
        if games_info != []:
            for games in range(len(games_info)):
                games_list.append(int(games_info[games].get_text()))
            games_data_list.append(games_list)

print(clubs_list)
print(games_data_list)
```

```python
EPL_dictionary = {}

for length in range(len(clubs_list)):
    EPL_dictionary[clubs_list[length]] = games_data_list[length]
    
print(EPL_dictionary)
```

```python
EPL_df = pd.DataFrame(EPL_dictionary, index = ['Games Played', 'Wins', 'Draws', 'Losses', 'Goals For', 'Goals Against', 
                                               'Goals difference', 'Points']).T
EPL_df.to_csv('EPL.csv')
EPL_df
```

```python
games_played_stats = EPL_df['Games Played'].describe()
points_stats = EPL_df['Points'].describe()

stats_dictionary = {'Games Played': games_played_stats, 'Points': points_stats}
stats_df = pd.DataFrame(stats_dictionary)
stats_df
```
